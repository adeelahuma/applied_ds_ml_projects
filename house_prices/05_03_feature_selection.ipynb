# Feature Selection on House Prices Data

- L1 Regularization

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

%matplotlib inline
```

```python
train_data = pd.read_pickle('data/final_train_data_feature_engineering_1hotenc_1.pkl')
train_data.info()
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
X = train_data[train_data.columns.drop("SalePrice_log")]
y = train_data["SalePrice_log"]

X.shape, y.shape
```

```python
X
```

## Normalize Data

```python
sc = StandardScaler()
```

```python
X.loc[: , X.columns] = sc.fit_transform(X.loc[:, X.columns])
```

## Test/Train split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Lasso Model

```python
lasso = Lasso(alpha=0.02, random_state=0).fit(X_train, y_train)
```

```python
model = SelectFromModel(lasso, prefit=True)
```

```python
X_train_new = model.transform(X_train)
X_train_new.shape
```

```python
X_train.shape
```

## Select  Features

```python
selected_features = pd.DataFrame(model.inverse_transform(X_train_new), 
             index = X_train.index, 
             columns = X_train.columns)
```

```python
selected_columns = selected_features.columns[selected_features.var()!=0]
selected_columns
```

## Save

```python
final_cols = selected_columns.tolist()
final_cols.append('SalePrice_log')
```

```python
pd.to_pickle(train_data[final_cols], 'data/final_train_data_l1_reg_v1.pkl')
```
