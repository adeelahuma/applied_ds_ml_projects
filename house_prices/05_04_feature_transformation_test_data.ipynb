## Adding new features


```python
def new_column_mapping(data):
    
    data['is_new_house'] = np.where(data['YearBuilt'] > 2000 , 1, 0)
    data['is_recent_remolding'] = np.where(data['YearRemodAdd'] > 2000 , 1, 0)
    data['basement_has_full_bathroom'] = np.where(data['BsmtFullBath'] > 0, 1, 0) 
    data['has_fireplace'] = np.where(data['Fireplaces'] > 0, 1, 0) 
    data['is_recent_garage_built'] = np.where(data['GarageYrBlt'] > 2000 , 1, 0)
    data['has_garage'] = np.where(data['GarageYrBlt'].isnull() , 0, 1)
    data['has_pool'] = np.where(data['PoolArea'] > 0, 1, 0)

    garage_area = np.where(data['GarageType'].isin(['Attchd', 'Basment', 'BuiltIn']), data['GarageArea'], 0)
    data['total_sqft_house'] = data['GrLivArea'] + garage_area + data['EnclosedPorch'] + data['TotalBsmtSF']

    # convert to categories
    for c in ['is_new_house', 'is_recent_remolding', 'basement_has_full_bathroom', 'has_fireplace',
                      'is_recent_garage_built', 'has_garage', 'has_pool']:
        data[c] = pd.Categorical(data[c])
```

```python
to_apply_transformation = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 
                           'TotalBsmtSF', '1stFlrSF', 'LowQualFinSF', 'GrLivArea',  
                           'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 
                           'PoolArea', 'MiscVal','total_sqft_house']

# to_apply_transformation_has_zero =  ['MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'TotalBsmtSF', 
#                                      'LowQualFinSF', 'BsmtHalfBath', 'KitchenAbvGr', 'WoodDeckSF', 
#                                      'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 
#                                      'PoolArea', 'MiscVal']
```

```python
def apply_transformation(data):
    
    for col in to_apply_transformation:

        col_name = col + '_log'
        if (data[col] < 0).sum() <= 0: ## to avoid log failure
            data[col_name] = np.log(data[col]+ 0.0001)
        else:    
            data[col_name] = np.log(data[col])

#     return data
```

## Binning 


```python
def binning(data):
    yr_built_bins = [1850, 1950, 1975, 2000, 2020]
    data['bin_yr_built'] = pd.cut(data['YearBuilt'], bins= yr_built_bins, precision=0)

    yr_remod_bins = [1950, 1980, 1990, 2000, 2020]
    data['bin_yr_remod'] = pd.cut(data['YearRemodAdd'], bins= yr_remod_bins, precision=0)

    yr_garage_bins = [1880, 1950, 1980, 1990, 2000,2020]
    data['bin_garage_built'] = pd.cut(data['GarageYrBlt'], bins= yr_garage_bins, precision=0)

    yr_sold_bins = [2006, 2007, 2008, 2009, 2010]
    data['bin_yr_sold'] = pd.cut(data['YrSold'], bins= yr_sold_bins, precision=0)

    month_sold_bins = [1, 4, 6, 9, 12]
    #month_sold_bins_lbl = ['spring', 'summer', 'fall', 'winter']
    data['bin_mo_sold'] = pd.cut(data['MoSold'], bins= month_sold_bins, precision=0)

#     return data
    
```

## Categorical Encoding 
- One-hot Encoding of all categorical variables with na dropped and kept 


```python
'''
    Convert given categorical columns to one-hot encoded columns
    drop 'NAN' column to avoid dummy variable trap
    drop categorical columns after 1-hot encoding 
    
    dummy_trap = True --> delete 1 dummy variable; 
        if column has null values then delete NA 
        if column has 1/0 value then delete 0 column
        
        Na and 0 signify lack of something 
'''

def get_one_hot_encoded_X(X, cat_colums, dummy_trap=True, delete_original_cat= True):

    for col in cat_colums:
        
        temp = pd.get_dummies(X[col], prefix=col, dummy_na= False if data[col].isnull().sum() == 0 else True)

        col_dummies = temp.columns
        
        if dummy_trap:
            col_dummies = [c for c in col_dummies if not( c.endswith('0') or c.endswith('nan'))]
        
        X = X.join(temp[col_dummies])

    if delete_original_cat:    
        X = X.drop(cat_colums, axis =1)   
        
    return X  
```
